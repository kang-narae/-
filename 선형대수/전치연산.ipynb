```python
#전치연산? 행과 열을 바꾼다.
#m행n열에 위치했던 값을 n행m열로 보냄.


import numpy as np
from sklearn.datasets import load_iris
iris=load_iris().data
iris




```

```python
x1= np.array([[1],[2],[3],[4],[5]])   #2차원의 열벡터
x2= np.array([[6],[7],[8],[9],[10]])
x3= np.array([[11],[12],[13],[14],[15]])  
```

```python
x1
```

```python
x2
```

```python
x3
```

```python
#하나의 데이터 레코드일 때는 열 벡터
#여러 데이터 레코드를 합칠 때는 열벡터를 행벡터로 바꾸고 붙이기로 했었지
'''
이걸 지난번에는 행벡터로 바꾸고 세로로 vstack으로 붙였었는데
이제는 한번에 전치
'''
x= np.hstack([x1,x2,x3])
x   #일단 열벡터를 reshape 없이 그대로 hstack으로 가로로 바꾸고 전치한다
```
